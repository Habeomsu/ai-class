```python
# Scale full set of predictors
X_scaled = StandardScaler().fit_transform(X)
```

```python
# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
# Scale train/ test predictors based on training data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        end_ix = i + n_steps
        if end_ix > len(sequence)-1:
            break
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# 윈도우 크기 설정 (입력 시퀀스의 길이)
window_size = 25

# 훈련 데이터셋과 테스트 데이터셋 준비
y_train_array = np.array(y_train)
train_set = np.c_[X_train_scaled,y_train_array]
y_test_array = np.array(y_test)
test_set = np.c_[X_test_scaled, y_test_array]

# 시퀀스 변환 함수 적용 (window_size = 25)
X_train_seq, y_train_seq = split_sequence(train_set, window_size)
X_test_seq, y_test_seq = split_sequence(test_set, window_size)
print(X_train_seq.shape, y_train_seq.shape)
print(X_test_seq.shape, y_test_seq.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
model.add(Dropout(0.1))
model.add(LSTM(units=50, activation='relu'))
model.add(Dense(units=1))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train_seq, y_train_seq, epochs=150, batch_size=160)
pred_y = model.predict(X_test_seq)
model.evaluate(X_test_seq, y_test_seq)
```
