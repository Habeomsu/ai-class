```python
from collections import Counter
import numpy as np

def split_sequences(sequences, n_steps):
    X, y = list(), list()  # ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•˜ì—¬ ì‹œí€€ìŠ¤ ë°ì´í„°ì™€ ë ˆì´ë¸”ì„ ë‹´ì„ ê³µê°„ì„ ë§Œë“¦
    for i in range(len(sequences)):  # ì „ì²´ ì‹œí€€ìŠ¤ ë°ì´í„°ë¥¼ ìˆœíšŒ
        # find the end of this pattern
        end_ix = i + n_steps  # í˜„ìž¬ ì¸ë±ìŠ¤(i)ì—ì„œ n_stepsë§Œí¼ ë–¨ì–´ì§„ ì‹œí€€ìŠ¤ì˜ ëì„ ê³„ì‚°
        # check if we are beyond the dataset
        if end_ix > len(sequences):  # ì‹œí€€ìŠ¤ ëì´ ë°ì´í„°ì˜ ë²”ìœ„ë¥¼ ë„˜ì–´ì„œëŠ”ì§€ í™•ì¸
            break  # ë²”ìœ„ë¥¼ ë„˜ìœ¼ë©´ ë£¨í”„ ì¢…ë£Œ
        # gather input (X) and output parts (y)
        seq_x = sequences[i:end_ix, :-1]  # ìž…ë ¥ ë°ì´í„° (íŠ¹ì§• ë°ì´í„°)
        seq_y_values = sequences[i:end_ix, -1]  # ì‹œí€€ìŠ¤ ë™ì•ˆì˜ ì¶œë ¥ ë°ì´í„° (ë ˆì´ë¸”ë“¤)
        
        # ê°€ìž¥ ë¹ˆë²ˆí•˜ê²Œ ë‚˜ì˜¨ ë ˆì´ë¸” ì°¾ê¸°
        most_common_label = Counter(seq_y_values).most_common(1)[0][0]
        
        X.append(seq_x)  # ìž…ë ¥ ë°ì´í„° ì¶”ê°€
        y.append(most_common_label)  # ê°€ìž¥ ë§Žì´ ë‚˜ì˜¨ ë ˆì´ë¸” ì¶”ê°€
    
    return np.array(X), np.array(y)  # ë¦¬ìŠ¤íŠ¸ë¥¼ numpy ë°°ì—´ë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜
```

```python
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Attention and Normalization
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(inputs, inputs)
    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    res = x + inputs

    # Feed Forward Part
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(res)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    return x + res
```

```python
def build_model(
    input_shape,
    head_size,
    num_heads,
    ff_dim,
    num_transformer_blocks,
    mlp_units,
    dropout=0,
    mlp_dropout=0,
):
    inputs = keras.Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    x = layers.GlobalAveragePooling1D(data_format="channels_last")(x)
    for dim in mlp_units:
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)
    outputs = layers.Dense(n_classes, activation="softmax")(x)
    return keras.Model(inputs, outputs)
```
